```python
import sqlite3
import pandas as pd

# Load CSV
df = pd.read_csv(
    '/home/dell/Downloads/carbon emission impact analysis/data/processed/final_dataset.csv'
)

# Create SQLite database
conn = sqlite3.connect(
    '/home/dell/Downloads/carbon emission impact analysis/data/processed/carbon.db'
)

# Store dataframe as SQL table
df.to_sql(
    'carbon_data',
    conn,
    if_exists='replace',
    index=False
)

print("Data loaded into SQLite successfully.")
```

```python
# Top 5 Sustainable Countries

query = """
SELECT country,
       AVG(renewable_energy) AS renewable,
       AVG(co2_per_capita) AS co2
FROM carbon_data
GROUP BY country
HAVING renewable > 50
ORDER BY co2 ASC
LIMIT 5;
"""

sustainable_countries = pd.read_sql(query, conn)

sustainable_countries
```

```python
# Countries With Highest Energy Consumption

query = """
SELECT country,
       AVG(energy_consumption) AS avg_energy
FROM carbon_data
GROUP BY country
ORDER BY avg_energy DESC
LIMIT 10;
"""

high_energy = pd.read_sql(query, conn)

high_energy
```

```python
# Countries With Highest Energy Consumption

query = """
SELECT country,
       AVG(energy_consumption) AS avg_energy
FROM carbon_data
GROUP BY country
ORDER BY avg_energy DESC
LIMIT 10;
"""

high_energy = pd.read_sql(query, conn)

high_energy
```

```python
# Year With Highest Average CO₂

query = """
SELECT year,
       AVG(co2_per_capita) AS avg_co2
FROM carbon_data
GROUP BY year
ORDER BY avg_co2 DESC
LIMIT 1;
"""

highest_co2_year = pd.read_sql(query, conn)

highest_co2_year
```

```python
# Average GDP By Decade

query = """
SELECT (year/10)*10 AS decade,
       AVG(gdp) AS avg_gdp
FROM carbon_data
GROUP BY decade
ORDER BY decade;
"""

gdp_decade = pd.read_sql(query, conn)

gdp_decade
```

```python
# Countries With Low Emissions But High GDP

query = """
SELECT country,
       AVG(gdp) AS avg_gdp,
       AVG(co2_per_capita) AS avg_co2
FROM carbon_data
GROUP BY country
HAVING avg_gdp > 100000000000
   AND avg_co2 < 5;
"""

low_emission_rich = pd.read_sql(query, conn)

low_emission_rich
```

```python
# Renewable Energy Leaders

query = """
SELECT country,
       MAX(renewable_energy) AS renewable_peak
FROM carbon_data
GROUP BY country
ORDER BY renewable_peak DESC
LIMIT 10;
"""

renewable_leaders = pd.read_sql(query, conn)

renewable_leaders
```
